```python
# Step 1: Install and Import Libraries
import subprocess
import sys

# Install required packages for CPU execution
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "transformers", "torch"])

print('✓ All libraries installed!')
print('Note: This demo runs on CPU. For GPU acceleration and full training, use Google Colab.')
```

```python
# Step 2: Create Sample Empathetic Dialogues Dataset
import pandas as pd

# Sample empathetic dialogue dataset
empathetic_dialogues = [
    {
        'context': 'I am feeling really sad and lonely today.',
        'response': 'I understand that you are going through a difficult time. It is perfectly normal to feel sad sometimes. Consider reaching out to friends or family, or speaking with a counselor. You are not alone.'
    },
    {
        'context': 'I have been having trouble sleeping lately.',
        'response': 'Sleep difficulties can be really challenging. Try establishing a bedtime routine, avoid screens before bed, and practice relaxation techniques like deep breathing. If it persists, consult a healthcare provider.'
    },
    {
        'context': 'I am struggling with anxiety about my future.',
        'response': 'Anxiety about the future is very common. Break down your concerns into manageable steps. Focus on what you can control today. Mindfulness and meditation can also help. Remember, you have overcome challenges before.'
    },
    {
        'context': 'I feel overwhelmed by my responsibilities.',
        'response': 'Feeling overwhelmed is a sign you might need to prioritize. List your tasks by importance and deadline. It is okay to ask for help or delegate. Take breaks to recharge. Your well-being comes first.'
    }
]

df = pd.DataFrame(empathetic_dialogues)
print('✓ Dataset created!')
print(f'Number of examples: {len(df)}')
print(f'\nFirst example:')
print(f'Context: {df["context"][0]}')
print(f'Response: {df["response"][0]}')
```

```python
# Step 3: Load the Model and Tokenizer
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = 'distilgpt2'  # Small, fast model
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# Load with map_location for CPU
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()  # Set to eval mode for inference

print('✓ Model loaded!')
print(f'Model: {model_name}')
print(f'Parameters: {sum(p.numel() for p in model.parameters()):,}')
print(f'Device: CPU (for GPU training, use Google Colab with GPU runtime)')
```

```python
# Step 4: Demonstrate Response Generation
from transformers import pipeline

generator = pipeline('text-generation', model=model, tokenizer=tokenizer, device=-1)

prompt = "User: I'm feeling overwhelmed and anxious about work.\nAssistant:"

response = generator(prompt, max_length=120, num_return_sequences=1, do_sample=True, temperature=0.7)
print('Prompt:')
print(prompt)
print('\nGenerated response:')
print(response[0]['generated_text'])

print('\nNote: This example demonstrates inference. Full fine-tuning requires a GPU-enabled environment.')
```

```python
# Step 5: Fine-Tune the Model
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='./mental_health_chatbot',
    num_train_epochs=2,
    per_device_train_batch_size=8,
    save_steps=500,
    logging_steps=100,
    learning_rate=5e-5,
    fp16=True,      # Faster training on GPU
    report_to='none'
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized,
)

print('Starting fine-tuning...')
trainer.train()
print('✓ Fine-tuning complete!')
```

```python
# Step 6: Test the Fine-Tuned Model
from transformers import pipeline

chatbot = pipeline('text-generation', model=model, tokenizer=tokenizer)

def mental_health_response(user_message):
    '''Generate empathetic response from fine-tuned model'''
    prompt = f'User: {user_message}\nResponse:'
    result = chatbot(prompt, max_new_tokens=80, do_sample=True,
                     temperature=0.7, pad_token_id=tokenizer.eos_token_id)
    generated = result[0]['generated_text']
    # Extract only the response part
    return generated.split('Response:')[-1].strip()

print('✓ Testing the fine-tuned model...\n')
```

```python
# Step 7: Demonstrate with Test Inputs
test_inputs = [
    'I feel really anxious today.',
    'I am stressed about work and cannot sleep.',
    'I feel lonely and disconnected from people.',
]

print('\n' + '=' * 70)
print('Mental Health Chatbot - Test Results')
print('=' * 70)

for msg in test_inputs:
    print(f'\nUser: {msg}')
    print('-' * 70)
    response = mental_health_response(msg)
    print(f'Bot: {response}')
    print('=' * 70)
```

## Important Notes

**Why Limited Performance?**
- We trained for only 2 epochs on 1000 examples for speed
- Production models train for many epochs on millions of examples
- This is sufficient to demonstrate the fine-tuning process

**What is Fine-Tuning?**
- Starting with a pre-trained model (distilgpt2)
- Training it further on a specific dataset (empathetic dialogues)
- Results in a model specialized for a specific task

**How to Improve:**
1. Train for more epochs (5-10)
2. Use more examples (5000-10000)
3. Use a larger base model (GPT-2 instead of DistilGPT2)
4. Implement evaluation metrics to track quality
